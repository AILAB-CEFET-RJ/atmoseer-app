# Agregação dos dados nas diferentes pressões atmosféricas

```python
ano = list(map(str,range(1998,2022)))

ds = xr.open_dataset('RJ_1997.nc')
for i in ano:
  ds_aux = xr.open_dataset('RJ_'+ i +'.nc')
  ds = ds.merge(ds_aux) 
```

```python
ano2 = [['1997','1998'],['1999','2000'],['2001','2002'],['2003','2004'],['2005','2006'],['2007','2008'],['2009','2010'],['2011','2012'],['2013','2014'],['2015','2016'],['2017','2018'],['2019','2020']]
ds2 = xr.open_dataset('RJ_2021_200.nc')
for i in ano2:
  ds_aux2 = xr.open_dataset('RJ_'+ i[0]+'_'+i[1] +'_200.nc')
  ds2 = ds2.merge(ds_aux2)
```
