```python
x = [2, 3, 4, 5, 12, 13, 14, 15, 16, 17, 22, 25, 26, 28, 51, 52, 57]

def group(L):
    first = last = L[0]
    for n in L[1:]:
        if n - 1 == last: # Part of the group, bump the end
            last = n
        else: # Not part of the group, yield current group and start a new
            yield first, last
            first = last = n
    yield first, last # Yield the last group


print(list(group(x)))
[(2, 5), (12, 17), (22, 22), (25, 26), (28, 28), (51, 52), (57, 57)]
```

```python
def find_contiguous_observation_blocks(df: pd.DataFrame):
    timestamp_range = df.index
    first = last = timestamp_range[0]
    for n in timestamp_range[1:]:
        previous = n - timedelta(hours=1, minutes=0)
        if previous == last: # Part of the current block, bump the end
            last = n
        else: # Not part of the current block; yield current block and start a new one
            yield first, last
            first = last = n
    yield first, last # Yield the last block


period_under_study = df_a652
# period_under_study = df_a652['2007-05-18':'2007-05-31']

contiguous_observation_blocks = list(find_contiguous_observation_blocks(period_under_study))
print(len(contiguous_observation_blocks))
print(contiguous_observation_blocks)
```

```python
contiguous_block_sizes = []
for p in list(group(df_a652['Datetime'])):
    start = p[0]
    end = p[1]
    diff = end - start
    diff_in_hours = diff.total_seconds() / 3600
    contiguous_block_sizes += [diff_in_hours]
print(contiguous_block_sizes)
```

```python
import matplotlib.pyplot as plt
# plotting labelled histogram
plt.hist(contiguous_block_sizes)
plt.xlabel('block size (in hours)')
plt.ylabel('block count')
plt.show()
```
